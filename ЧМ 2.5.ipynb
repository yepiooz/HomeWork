```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return x**3 - x - 1

def df(x):
    return 3*x**2 - 1

# Метод хорд
def secant_method(x0, x1, n_iter):
    xs = [x0, x1]
    for _ in range(n_iter - 1):
        x_prev, x_curr = xs[-2], xs[-1]
        fx_prev, fx_curr = f(x_prev), f(x_curr)
        x_next = x_curr - fx_curr * (x_curr - x_prev) / (fx_curr - fx_prev)
        xs.append(x_next)
    return xs

# Метод Ньютона
def newton_method(x0, n_iter):
    xs = [x0]
    for _ in range(n_iter):
        x_curr = xs[-1]
        x_next = x_curr - f(x_curr) / df(x_curr)
        xs.append(x_next)
    return xs

# Параметры
a, b = -1, 2
X = np.linspace(a, b, 400)
Y = f(X)

# ---------- График метода хорд ----------
x0, x1 = 0.0, 1.5   
sec_points = secant_method(x0, x1, 6)

plt.figure(figsize=(6, 5))
plt.plot(X, Y, label='y = f(x)')
plt.axhline(0, color='black', linewidth=0.8)


for k in range(1, len(sec_points)):
    x_prev, x_curr = sec_points[k-1], sec_points[k]
    plt.plot([x_prev, x_curr], [f(x_prev), f(x_curr)], 'r-')
    plt.plot(x_curr, 0, 'ko')

plt.title('Метод хорд')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# ---------- График метода Ньютона ----------
newt_points = newton_method(1.5, 5)

plt.figure(figsize=(6, 5))
plt.plot(X, Y, label='y = f(x)')
plt.axhline(0, color='black', linewidth=0.8)


for k in range(len(newt_points) - 1):
    xk = newt_points[k]
    fxk = f(xk)
    tX = np.linspace(xk - 1, xk + 1, 20)
    tY = df(xk) * (tX - xk) + fxk
    plt.plot(tX, tY, 'r--')     
    plt.plot(newt_points[k+1], 0, 'ko')  

plt.title('Метод касательных (Ньютона)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()


```
